```python
# Clean Best Books Ever
best_books_cleaned = best_books.rename(columns={
    'author': 'author',
    'rating': 'average_rating',
    'description': 'description',
    'genres': 'genres',
    'publishDate': 'published_year',
    'pages': 'num_pages',
    'numRatings': 'ratings_count'
})[['title', 'author', 'genres', 'description', 'published_year', 'average_rating', 'num_pages', 'ratings_count']]

# Extract year from publishDate (e.g., '9/14/2008' ➔ 2008)
best_books_cleaned['published_year'] = pd.to_datetime(best_books_cleaned['published_year'], errors='coerce').dt.year

# Clean 7k Books Metadata
books_metadata_cleaned = books_metadata.rename(columns={
    'authors': 'author',
    'categories': 'genres'
})[['title', 'author', 'genres', 'description', 'published_year', 'average_rating', 'num_pages', 'ratings_count']]
```

```python
# Merge both cleaned datasets
books = pd.concat([best_books_cleaned, books_metadata_cleaned], ignore_index=True)

# Drop duplicates based on title
books = books.drop_duplicates(subset='title')
```

```python
# Fill missing values
books['author'] = books['author'].fillna('Unknown Author')
books['genres'] = books['genres'].fillna('Unknown')
books['description'] = books['description'].fillna('No Description Available')

books['published_year'] = books['published_year'].fillna(0).astype(int)
books['average_rating'] = books['average_rating'].fillna(0)

#Fix num_pages safely
books['num_pages'] = (
    books['num_pages']
    .astype(str)
    .str.extract('(\d+)')  # extract only digits
    .fillna(0)
    .astype(int)
)

books['ratings_count'] = books['ratings_count'].fillna(0).astype(int)
# Save final dataset
books.to_csv("books_final_merged.csv", index=False)
```
