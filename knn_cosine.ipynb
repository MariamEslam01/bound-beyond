```python
#Load cleaned datasets (books and movies)
import pandas as pd

books = pd.read_csv("books_cleaned2.csv")
movies = pd.read_csv("movies_cleaned.csv")
```

```python
# Rename columns
books = books.rename(columns={'description': 'content'})
movies = movies.rename(columns={'overview': 'content'})
```

```python
# For books
books['type'] = 'book'
books['image_url'] = None  # No thumbnail available
books_subset = books[['title', 'content', 'author', 'published_year', 'average_rating', 'genres', 'image_url', 'type']].copy()

# For movies
base_movie_poster_url = "https://image.tmdb.org/t/p/w500"
movies['type'] = 'movie'
movies['content'] = movies['tagged_Movie_description']
movies['image_url'] = base_movie_poster_url + movies['poster_path']
movies_subset = movies[['title', 'content', 'published_year', 'average_rating', 'genres', 'image_url', 'type']].copy()
movies_subset['author'] = None

# Combine
combined = pd.concat([books_subset, movies_subset], ignore_index=True)
combined['content'] = combined['content'].fillna('')

```

```python
def format_book_output(row):
    author = row.get('author', 'Unknown Author')
    published_year = row.get('published_year', 'Unknown Year')
    average_rating = row.get('average_rating', 0)

    return f"📚 {row['title']} by {author} ({published_year}) ⭐ {average_rating:.2f}"

def format_movie_output(row):
    release_year = row.get('published_year', 'Unknown Year')
    average_rating = row.get('average_rating', 0)

    if isinstance(release_year, (int, float)):
        year_display = int(release_year)
    else:
        year_display = 'Unknown Year'

    return f"🎥 {row['title']} ({year_display}) ⭐ {average_rating:.2f}"
```
